# **Fundamentals of Linear Algebra- the Language of Data**

To understand how neurons respond to different stimuli, we can measure their firing rates and analyse the patterns using vectors. We apply a stimulus and record their firing rates:

-	Neuron A: 20 Hz
-	Neuron B: 15 Hz

We then apply a different stimulus and record again:

-	Neuron A: 5 Hz
-	Neuron B: 10 Hz


By repeating this experiment with various stimuli, we gather a large amount of data. However, to analyse this data effectively, we need a structured way to record it. This is where vectors become useful.

---


## **1. Definition of Vectors**

A **vector** is an ordered list of numbers, where the order matters. You can also think of a vector as an arrow in space, with both direction and magnitude.
For example, the two recordings above can be represented as column vectors:

First observation:

$$
\mathbf{v_1} = \begin{pmatrix} 20 \\ 15 \end{pmatrix}
$$

Second observation:

$$
\mathbf{v_2} = \begin{pmatrix} 5 \\ 10 \end{pmatrix}
$$

In these vectors, the first element corresponds to neuron A’s firing rate, and the second to neuron B’s. Each vector represents the neural response to a specific stimulus.
Each vector captures a response to one stimulus. By collecting vectors for multiple stimuli, we can assemble them into a matrix for more comprehensive analysis, e.g.
$$
\mathbf{V} = \begin{pmatrix} \mathbf{v_1} \ \mathbf{v_2} \end{pmatrix} = \begin{pmatrix} 20 & 5 \\ 15 &  10  \end{pmatrix}
$$

In this matrix **V**, each column corresponds to a different stimulus, and each row denotes the firing frequency of a particular neuron.

This matrix can be extended to include the observations of *m* neurons, for *n* different stimuli:


$$
\mathbf{V} = \begin{pmatrix}
v_{11} & v_{12} & \cdots & v_{1n} \\
v_{21} & v_{22} & \cdots & v_{2n} \\
\vdots & \vdots & \ddots & \vdots \\
v_{m1} & v_{m2} & \cdots & v_{mn}
\end{pmatrix}
$$

```python
import numpy as np

# Define the vectors as column vectors (2x1 shape)
v1 = np.array([[20], [15]])
v2 = np.array([[5], [10]])

# Combine into a matrix (each column is a vector)
V = np.hstack([v1, v2])

# Display the result
print("Matrix V:\n", V)
```


## **2. Vector Arithmetic**

### Vector Addition

To add two vectors, you simply add their corresponding components. For example, if:
$$ \mathbf{u} = \begin{pmatrix} u_1 \ u_2 \end{pmatrix} $$ and $$ \mathbf{v} = \begin{pmatrix} v_1 \ v_2 \end{pmatrix} $$

Then:
$$ \mathbf{u} + \mathbf{v} = \begin{pmatrix} u_1 + v_1 \ u_2 + v_2 \end{pmatrix} $$

---


### Vector Subtraction

To subtract two vectors, you subtract their corresponding components. For example:
$$ \mathbf{u} - \mathbf{v} = \begin{pmatrix} u_1 - v_1 \ u_2 - v_2 \end{pmatrix} $$


---

### Scalar Multiplication

To multiply a vector by a scalar, you multiply each component of the vector by the scalar. For example, if: c is a scalar and $$ \mathbf{v} = \begin{pmatrix} v_1 \ v_2 \end{pmatrix} $$

Then:
$$ c \mathbf{v} = \begin{pmatrix} c v_1 \ c v_2 \end{pmatrix} $$

---

### Magnitude (Length) and Direction of a Vector
Given a 2D vector:
$$ \mathbf{v} = \begin{pmatrix} v_1 \ v_2 \end{pmatrix} $$


The **magnitude** (or length) of a vector $ \mathbf{v} $  is given by the Euclidean norm:

$$
|\mathbf{v}| = \sqrt{v_1^2 + v_2^2}
$$


The **direction** (angle  $\theta$ ) of the vector relative to the positive x-axis is:

$$
\theta = \tan^{-1} \left( \frac{v_1}{v_2} \right)
$$

To convert the angle from radians to degrees:

$$
\theta_{\text{deg}} = \theta \times \left( \frac{180}{\pi} \right)
$$

---

```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Define vectors
v_1 = np.array([20, 15])
v_2 = np.array([5, 10])

# Vector arithmetic
v_add = v_1 + v_2
v_sub = v_1 - v_2

# Magnitude function
def magnitude(v):
    return np.linalg.norm(v)

# Direction (angle in degrees)
def direction_deg(v):
    angle_rad = math.atan2(v[1], v[0])
    return math.degrees(angle_rad)

# Compute magnitudes
mag_v1 = magnitude(v_1)
mag_v2 = magnitude(v_2)

# Compute directions
dir_v1 = direction_deg(v_1)
dir_v2 = direction_deg(v_2)

# Print results
print("Vector v₁:", v_1)
print("Vector v₂:", v_2)
print("v₁ + v₂ =", v_add)
print("v₁ - v₂ =", v_sub)
print(f"Magnitude of v₁: {mag_v1:.2f}")
print(f"Magnitude of v₂: {mag_v2:.2f}")
print(f"Direction of v₁: {dir_v1:.2f}°")
print(f"Direction of v₂: {dir_v2:.2f}°")

# Plotting
origin = np.array([0, 0])  # origin point

plt.figure(figsize=(8, 8))
plt.quiver(*origin, *v_1, color='r', angles='xy', scale_units='xy', scale=1, label='v₁')
plt.quiver(*origin, *v_2, color='b', angles='xy', scale_units='xy', scale=1, label='v₂')
plt.quiver(*origin, *v_add, color='g', angles='xy', scale_units='xy', scale=1, label='v₁ + v₂')
plt.quiver(*origin, *v_sub, color='purple', angles='xy', scale_units='xy', scale=1, label='v₁ - v₂')

# Plot settings
plt.xlim(0, 30)
plt.ylim(0, 30)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.gca().set_aspect('equal')
plt.legend()
plt.title("Vector Arithmetic: v₁, v₂, v₁+v₂, v₁−v₂")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```


### Unit Vectors and Normalization

A **unit vector** has a magnitude of exactly 1 and points in the *same direction as the original vector*. It represents **direction only**, without considering length.

To convert a vector $\mathbf{v}$ into a unit vector $\hat{\mathbf{v}}$, use:

$$
\hat{\mathbf{v}} = \frac{\mathbf{v}}{|\mathbf{v}|}
$$

This process is called **normalization**.

```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Define vectors
v_1 = np.array([20, 15])
v_2 = np.array([5, 10])

def normalize(v):
    norm = np.linalg.norm(v)
    if norm == 0:
        return v
    return v / norm

# Normalize v1 and v2
unit_v1 = normalize(v_1)
unit_v2 = normalize(v_2)

# Check magnitudes
mag_unit_v1 = np.linalg.norm(unit_v1)
mag_unit_v2 = np.linalg.norm(unit_v2)

print("Unit vector of v₁:", unit_v1)
print("Unit vector of v₂:", unit_v2)
print(f"Check magnitude of unit v₁: {mag_unit_v1:.2f}")
print(f"Check magnitude of unit v₂: {mag_unit_v2:.2f}")

# Plot original and unit vectors
plt.figure(figsize=(6, 6))

# Original vectors (faint)
plt.quiver(*origin, *v_1, color='r', alpha=0.3, angles='xy', scale_units='xy', scale=1, label='Original v₁')
plt.quiver(*origin, *v_2, color='b', alpha=0.3, angles='xy', scale_units='xy', scale=1, label='Original v₂')

# Unit vectors (bright)
plt.quiver(*origin, *unit_v1, color='r', angles='xy', scale_units='xy', scale=1, label='Unit v₁')
plt.quiver(*origin, *unit_v2, color='b', angles='xy', scale_units='xy', scale=1, label='Unit v₂')

# Axes and formatting
plt.xlim(-0.2, 20.2)
plt.ylim(-0.2, 15.2)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.gca().set_aspect('equal')
plt.legend()
plt.title("Original vs Unit Vectors")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

## **3. Comparing Neural Responses: Similarity and Alignment**

As explained previously, in neuroscience, the activity of a population of neurons in response to a stimulus can be represented as a **vector**:

- Each **dimension** corresponds to one neuron
- Each **vector** represents the firing rates of all neurons in response to a **single stimulus**

Let’s define two such response vectors:

- **Stimulus A** evokes response $\mathbf{v}_1 = \begin{bmatrix} 20 \\ 15 \end{bmatrix}$
- **Stimulus B** evokes response $\mathbf{v}_2 = \begin{bmatrix} 5 \\ 10 \end{bmatrix}$

---

### Dot Product

The **dot product** of two vectors is defined as:

$$
\mathbf{v}_1 \cdot \mathbf{v}_2 = v_{1x} \cdot v_{2x} + v_{1y} \cdot v_{2y}
$$

More generally:

$$
\mathbf{v}_1 \cdot \mathbf{v}_2 = \sum_{i=1}^n v_{1i} \cdot v_{2i}
$$

This gives a scalar value representing how much the **population response to Stimulus A overlaps with that of Stimulus B**.

```python
import numpy as np
v1 = np.array([20, 15])
v2 = np.array([5, 10])
dot_product = np.dot(v1, v2)
print(f"Dot product: {dot_product}")
```

### Cosine Similarity

To compare the **direction** (not magnitude) of two stimulus-evoked neural response vectors, we use **cosine similarity**:

$$
\cos(\theta) = \frac{\mathbf{v}_1 \cdot \mathbf{v}_2}{\|\mathbf{v}_1\| \cdot \|\mathbf{v}_2\|}
$$

- $\mathbf{v}_1$, $\mathbf{v}_2$ are the neural responses to two different stimuli.
- $\|\mathbf{v}_1\|$ and $\|\mathbf{v}_2\|$ are the Euclidean norms (vector lengths).
- $\theta$ is the angle between the two vectors.
- $\cos(\theta)$ ranges from:
  - **1** → perfectly aligned (same direction)
  - **0** → orthogonal (no shared direction)
  - **–1** → opposite directions


```python
import numpy as np

v1 = np.array([20, 15])  # response to stimulus A
v2 = np.array([5, 10])   # response to stimulus B

dot_product = np.dot(v1, v2)
mag_v1 = np.linalg.norm(v1)
mag_v2 = np.linalg.norm(v2)

cos_theta = dot_product / (mag_v1 * mag_v2)
angle_deg = np.degrees(np.arccos(cos_theta))

print(f"Cosine similarity: {cos_theta:.3f}")
print(f"Angle between vectors: {angle_deg:.2f}°")
```

### Projection

The **projection** of one vector onto another gives the component of the first vector that lies **in the direction** of the second vector.

In our neuroscience context:

- $\mathbf{v}_1$ and $\mathbf{v}_2$ are firing rate vectors (population responses to two different stimuli).
- The projection of $\mathbf{v}_1$ onto $\mathbf{v}_2$ shows **how much of the response to stimulus A aligns with the pattern evoked by stimulus B**.

The formula for the projection of $\mathbf{v}_1$ onto $\mathbf{v}_2$ is:

$$
\mathrm{proj}_{\mathbf{v}_2}(\mathbf{v}_1) = \left( \frac{\mathbf{v}_1 \cdot \mathbf{v}_2}{\|\mathbf{v}_2\|^2} \right) \mathbf{v}_2
$$

- $\mathbf{v}_1 \cdot \mathbf{v}_2$ is the dot product
- $\|\mathbf{v}_2\|$ is the magnitude (Euclidean norm) of $\mathbf{v}_2$
- The result is a vector **in the direction of** $\mathbf{v}_2$


```python
import numpy as np

v1 = np.array([20, 15])  # Response to Stimulus A
v2 = np.array([5, 10])   # Response to Stimulus B

dot_product = np.dot(v1, v2)
projection = (dot_product / np.dot(v2, v2)) * v2
print(f"Projection of v1 onto v2: {projection}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 6))
plt.quiver(0, 0, v1[0], v1[1], color='r', label='Stimulus A (v1)', angles='xy', scale_units='xy', scale=1)
plt.quiver(0, 0, v2[0], v2[1], color='b', label='Stimulus B (v2)', angles='xy', scale_units='xy', scale=1)
plt.quiver(0, 0, projection[0], projection[1], color='g', label='projᵥ₂(v1)', angles='xy', scale_units='xy', scale=1)
plt.xlim(0, 25)
plt.ylim(0, 20)
plt.grid()
plt.gca().set_aspect('equal')
plt.legend()
plt.title("Projection of Stimulus A onto Stimulus B")
plt.xlabel("Neuron 1 firing rate")
plt.ylabel("Neuron 2 firing rate")
plt.show()
```

## **4. Linear Independence**

In the previous plots, we created new vectors by adding and subtracting two vectors, $\mathbf{v}_1 $ and $\mathbf{v}_2 $. These are examples of **linear combinations**, which involve forming a new vector by scaling and then adding other vectors together. Mathematically, a linear combination of $\mathbf{v}_1$ and $\mathbf{v}_2$ is expressed as:

$$
\mathbf{v} = a \mathbf{v}_1 + b \mathbf{v}_2
$$

where $(a, b)$ are scalars (real numbers), i.e. $ a,b \in \mathbb{R}$

Examples:

$$
\mathbf{v} = \mathbf{v}_1 + \mathbf{v}_2, \quad a = b = 1
$$

$$
\mathbf{v} = \mathbf{v}_1 - \mathbf{v}_2, \quad a = 1, b = -1
$$

In the context of vectors, **linear independence** refers to a situation where no vector in a set can be written as a combination of the others. Specifically, a set of vectors $\{\mathbf{v}_1, \mathbf{v}_2, \ldots, \mathbf{v}_n\}$ is said to be linearly independent if the only solution to the equation

$$
a_1 \mathbf{v}_1 + a_2 \mathbf{v}_2 + \cdots + a_k \mathbf{v}_k = \mathbf{0}
$$

is when all the scalar coefficients $ a_1, a_2, \ldots, a_n $ are equal to zero (i.e., $(a_1 = a_2 = \cdots = a_k = 0)$).

In other words, no vector in the set can be "recreated" by combining the others. If a non-zero solution to the equation exists (i.e., some $a_i$ are not zero), then the set of vectors is *linearly dependent*.



```python
import numpy as np
import matplotlib.pyplot as plt

# --- Function to check linear dependence ---
def is_linearly_dependent(vectors):
    mat = np.column_stack(vectors)
    rank = np.linalg.matrix_rank(mat)
    return rank < len(vectors)

# --- Example 1: Linearly Independent Vectors ---
v1 = np.array([2, 3])
v2 = np.array([-1, 4])

# --- Example 2: Linearly Dependent Vectors ---
d1 = np.array([2, 3])
d2 = 3 * d1  # Scalar multiple

# --- Plotting Function ---
def plot_vectors(vecs, colors, title):
    origin = np.array([0, 0])
    plt.figure(figsize=(6, 6))
    for v, c in zip(vecs, colors):
        plt.quiver(*origin, *v, angles='xy', scale_units='xy', scale=1, color=c)
    plt.axhline(0, color='gray', linewidth=0.5)
    plt.axvline(0, color='gray', linewidth=0.5)
    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
    plt.gca().set_aspect('equal')
    plt.grid(True)
    plt.title(title)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.show()

# --- Check and plot ---
print("Set 1: v1 and v2")
print("Linearly Dependent?" , is_linearly_dependent([v1, v2]))
plot_vectors([v1, v2], ['blue', 'orange'], "Linearly Independent Vectors")

print("Set 2: d1 and d2")
print("Linearly Dependent?" , is_linearly_dependent([d1, d2]))
plot_vectors([d1, d2], ['green', 'red'], "Linearly Dependent Vectors")
```

## **5. Introduction to Basis Vectors**

A **basis** of a vector space is a set of vectors that are:

- **Linearly independent**, and
- **Span** the entire vector space.

Every vector in the space can be uniquely expressed as a **linear combination** of the basis vectors.

---

### Basis Vectors in 2D

The most common example is the standard basis for the 2D vector space $\mathbb{R}^2$:

$$
\mathbf{e}_1 = \begin{pmatrix} 1 \\ 0 \end{pmatrix}, \quad
\mathbf{e}_2 = \begin{pmatrix} 0 \\ 1 \end{pmatrix}
$$

Any vector $\mathbf{v} \in \mathbb{R}^2$ can be written as:

$$
\mathbf{v} = v_1 \mathbf{e}_1 + v_2 \mathbf{e}_2 = \begin{pmatrix} v_1 \\ v_2 \end{pmatrix}
$$

---

### 3D Vectors in a 2D Vector Space

Sometimes, we deal with vectors that have three components but **lie in a 2D subspace** of $\mathbb{R}^3$. For example, consider vectors in $\mathbb{R}^3$ where the third component is always zero:

$$
\mathbf{v} = \begin{pmatrix} v_1 \\ v_2 \\ 0 \end{pmatrix}
$$

A basis for this 2D subspace embedded in 3D is:

$$
\mathbf{e}_1 = \begin{pmatrix} 1 \\ 0 \\ 0 \end{pmatrix}, \quad
\mathbf{e}_2 = \begin{pmatrix} 0 \\ 1 \\ 0 \end{pmatrix}
$$

Any vector in this subspace can be expressed as:

$$
\mathbf{v} = v_1 \mathbf{e}_2
$$


```python
import matplotlib.pyplot as plt
import numpy as np

# Define two linearly independent vectors in 3D
v1 = np.array([1, 2, 1])
v2 = np.array([2, -1, 0])

# Define a redundant vector as a linear combination of v1 and v2
v3 = 1.5 * v1 - 0.5 * v2

# Create a grid of linear combinations of v1 and v2
a = np.linspace(-2, 2, 20)
b = np.linspace(-2, 2, 20)
A, B = np.meshgrid(a, b)
X = A * v1[0] + B * v2[0]
Y = A * v1[1] + B * v2[1]
Z = A * v1[2] + B * v2[2]

# Set up the 3D plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot the plane (span of v1 and v2)
ax.plot_surface(X, Y, Z, alpha=0.5, color='lightblue', edgecolor='gray')

# Plot basis vectors v1 and v2
origin = np.zeros(3)
ax.quiver(*origin, *v1, color='red', label=r'$\mathbf{v}_1$', linewidth=2)
ax.quiver(*origin, *v2, color='green', label=r'$\mathbf{v}_2$', linewidth=2)

# Plot the redundant vector v3
ax.quiver(*origin, *v3, color='purple', label=r'$\mathbf{v}_3 = 1.5\mathbf{v}_1 - 0.5\mathbf{v}_2$', linewidth=2)

# Set plot limits for clarity
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.set_zlim(-3, 5)

# Labels, title, and legend
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('2D Subspace (Plane) in 3D Spanned by $\mathbf{v}_1$ and $\mathbf{v}_2$ with Dependent Vector $\mathbf{v}_3$')
ax.legend()

plt.tight_layout()
plt.show()

```

A commonly used basis is the standard basis for 3D space:

$$
\mathbf{e}_1 = \begin{pmatrix} 1 \\ 0 \\ 0 \end{pmatrix}, \quad
\mathbf{e}_2 = \begin{pmatrix} 0 \\ 1 \\ 0 \end{pmatrix}, \quad
\mathbf{e}_3 = \begin{pmatrix} 0 \\ 0 \\ 1 \end{pmatrix}
$$

These are unit vectors along the $x$-, $y$-, and $z$-axes, respectively. Any vector $\mathbf{v} = \begin{pmatrix} x \\ y \\ z \end{pmatrix}$ in 3D can be written as:

$$
\mathbf{v} = x \mathbf{e}_1 + y \mathbf{e}_2 + z \mathbf{e}_3
$$



## **6. Dimensionality of Vector Space**

The concept of a basis is fundamental to understanding the dimensionality of a vector space.

A basis is a set of vectors that are linearly independent and span the entire vector space. For example, in 3D space, the standard basis vectors $\mathbf{e}_1$, $\mathbf{e}_2$, and $\mathbf{e}_3$ each point along one of the coordinate axes. Any vector $\mathbf{v}$ in this space can be uniquely expressed as a linear combination of these basis vectors:

$$
\mathbf{v} = x\mathbf{e}_1+y\mathbf{e}_2+z\mathbf{e}_3
$$

Here, $x$, $y$, and $z$ are scalars representing the coordinates of $\mathbf{v}$ in the basis. Since the basis consists of three vectors, we say that the vector space is 3-dimensional.

More generally, the dimension of a vector space is the number of vectors in any of its bases. This number tells us the minimum number of coordinates needed to represent any vector in the space. For instance, a 2D subspace of $\mathbb{R}^3$ has dimension 2 because it can be spanned by two linearly independent vectors, even though each vector has three components.

The rank of a set of vectors (or a matrix formed by these vectors) is the maximum number of linearly independent vectors in that set. Equivalently, rank measures the dimension of the subspace spanned by the vectors. For example, if three vectors in $\mathbb{R}^3$ have rank 2, it means only two of them are linearly independent, and all three lie within a 2-dimensional subspace (a plane).

Understanding dimensionality and rank helps us analyze the structure and complexity of vector spaces, which is essential in fields like physics, computational neuroscience, and data science.


```python
import numpy as np

# Define firing patterns of 3 neurons across different stimuli or times
# Each vector represents firing rates of 3 neurons

v1 = np.array([1,8,4])
v2 = np.array([5,7,2])
v3 = np.array([45,162,72])  # This is  a combination of v1 and v2 (linear dependence)

# Stack as matrix (each column a vector)
patterns = np.column_stack((v1, v2, v3))

print("Firing patterns matrix (each column a vector):")
print(patterns)

# Check rank to determine linear independence
rank = np.linalg.matrix_rank(patterns)
print(f"\nRank of the firing patterns matrix: {rank}")

# Number of vectors
num_vectors = patterns.shape[1]

if rank == num_vectors:
    print("The firing patterns are linearly independent — all carry unique information.")
else:
    print("The firing patterns are linearly dependent — some patterns are redundant.")

# Example: Express pattern3 as a linear combination of pattern1 and pattern2
# Solve: a*pattern1 + b*pattern2 = pattern3
coeffs = np.linalg.lstsq(patterns[:, :2], v3, rcond=None)[0]
print(f"\nCoefficients to express v3 as combination of v1 and v2: a={coeffs[0]:.2f}, b={coeffs[1]:.2f}")
print(f"\n i.e. v3 = {coeffs[0]:.2f}v1 + {coeffs[1]:.2f}v2")
```

In theory, if you measure data across $n$ neurons, each observation can be considered a point in $n$-dimensional space ($\mathbb{R}^n$). For example:

- If you record the firing rates of 3 neurons, each observation is a point in 3D space.
- If you record from 100 neurons, the data lives in 100-dimensional space.

But just because the data could exist in $\mathbb{R}^n$ doesn’t mean it actually fills that whole space. In practice, neural data often lies in a lower-dimensional subspace.

Example:

Imagine recording from 50 neurons in the motor cortex while a monkey moves its arm in different directions. You might find:

- The data lives in a 3D subspace (e.g., corresponding to $x$-, $y$-, $z$- movement directions).
- Despite having 50 neurons, the brain is controlling movement using only a few coordinated activity patterns.

It means that although each data point is recorded in a high-dimensional space, the actual variability in the data may only occupy a small part of that space. This low-dimensional control is often called a **neural manifold**.

The number of linearly independent vectors in a space is the **dimension** of that space.

So, if you have 100-dimensional data, but only 3 linearly independent vectors among them, your data effectively lies in 3D, not 100D.

These 3 independent vectors form a **basis** — any other vector in the dataset can be written as a linear combination of them.

For example, in neural recordings:

- Each observation is a vector of firing rates: $\mathbf{v} \in \mathbb{R}^n$.

- If you collect multiple such observations, you’re forming a set of vectors in $\mathbb{R}^n$.

- The number of linearly independent vectors tells you the true dimensionality of the data.

This is why when researchers discover that 100 neurons’ activity only spans 3--10 dimensions, they’re saying: most of the recorded vectors are linear combinations of just a few independent activity patterns.



## **7. Eigenvalues**

An **eigenvector** of a matrix or system can be thought of as a vector that points in a specific direction where, when the system or matrix is applied to it, the vector only scales (stretches or shrinks) but does not change direction.

**Example:** Imagine a dataset of points in space. An eigenvector can be thought of as the “most significant direction” along which the points are spread out. In the context of neural data, this might correspond to the most significant pattern or trend in how neurons fire together, where the eigenvector represents the *direction* of greatest variance or activity.

*Eigenvectors* point in the direction of greatest variance in the data (or the most important direction in a dataset). We can think of the eigenvector as the direction in which the data is most "spread out."

*Eigenvalues* tell us how much the data is stretched or compressed along that direction. Specifically:

- If the eigenvalue is large, it means the data is stretched a lot along that eigenvector direction. The points are more spread out along the eigenvector.

- If the eigenvalue is small (close to 0), it means the data is compressed along that eigenvector. The points are tightly packed in that direction.

- If the eigenvalue is zero, it means there is no spread in that direction—i.e., all data points are clustered along a lower-dimensional subspace (like a flat line or point).




```python
import numpy as np
import matplotlib.pyplot as plt

# Seed for reproducibility
np.random.seed(0)

# Generate 2D data mostly along a direction but with noise
n_points = 300
direction = np.array([2, 1])  # arbitrary direction
direction = direction / np.linalg.norm(direction)  # normalize

# Create data points scaled along the direction plus some perpendicular noise
scales = np.random.normal(loc=0, scale=5, size=n_points)
noise = np.random.normal(loc=0, scale=1, size=n_points)  # small noise perpendicular

# Perpendicular vector
perp_direction = np.array([-direction[1], direction[0]])

# Generate dataset
data = np.outer(scales, direction) + np.outer(noise, perp_direction)

# Compute covariance matrix
cov_matrix = np.cov(data.T)

# Calculate eigenvalues and eigenvectors
eigvals, eigvecs = np.linalg.eigh(cov_matrix)

# Sort eigenvalues and eigenvectors in descending order
idx = np.argsort(eigvals)[::-1]
eigvals = eigvals[idx]
eigvecs = eigvecs[:, idx]

# Plot the data points
plt.figure(figsize=(8, 6))
plt.scatter(data[:, 0], data[:, 1], alpha=0.5, label='Data points')

# Plot eigenvectors scaled by eigenvalues for visualization
origin = np.mean(data, axis=0)
for i in range(2):
    vec = eigvecs[:, i] * np.sqrt(eigvals[i]) * 3  # scale for visibility
    plt.quiver(*origin, *vec, angles='xy', scale_units='xy', scale=1,
               label=f'Eigenvector {i+1} (eigval={eigvals[i]:.2f})', width=0.005)

plt.axis('equal')
plt.title('2D Data with Eigenvectors and Eigenvalues')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()

print("Eigenvalues:", eigvals)
print("Eigenvectors (columns):\n", eigvecs)
```

## **Applications to Neural Data**
Neural data can be analysed using these linear algebra concepts to understand the underlying patterns of neural activity. Identifying the most important directions in high-dimensional neural data can help reduce complexity (through dimensionality reduction) and reveal the *core structure* of brain activity.


| Concept          | Mathematical Meaning                           | Neuroscience/Data Science Context                     |
|------------------|-----------------------------------------------|-------------------------------------------------------|
| Vector           | Element of $\mathbb{R}^n$                      | A single neural activity pattern                       |
| Dot Product      | $\mathbf{v}_1 \cdot \mathbf{v}_2 = \sum_{i=1}^n v_{1i} \cdot v_{2i}$      | Measure of similarity between patterns          |
| Linear combination| $a \mathbf{v}_1 + b \mathbf{v}_2 + \cdots$    | Combining neural patterns or responses                 |
| Span             | Set of all linear combinations                 | The set of possible responses the brain can generate  |
| Basis            | Minimal spanning set of linearly independent vectors | Fundamental patterns underlying the data           |
| Dimension        | Number of vectors in a basis                    | Degrees of freedom in the neural data                  |
| Rank of matrix   | Dimension of column space                        | Effective dimensionality of recorded activity          |
| Manifold         | Nonlinear lower-dimensional surface             | Curved, structured data patterns in high-dimensional space |
